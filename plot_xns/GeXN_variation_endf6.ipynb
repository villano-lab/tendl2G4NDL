# Reading Ge Variations with ENDF6 Library

What I'm trying to do here is read the ENDF6 files that Danika got from Dr. Rochman (email dated 4/5/18) so that we can use them to bound the variation of the spectrum that is expected in the photoneutron analysis.

```python
import pickle
from pylab import *
import matplotlib.image as mpimg
from matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import h5py
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import gaussian_kde
import ENDF6
```

```python
#get some cross section data
f = open('/data/chocula/villaa/PhotoN_SuperSim/cross_sections/neutron_file/Ge/Ge070/lib/endf/random/n-Ge070-rand-0000')
lines = f.readlines()
sec = ENDF6.find_section(lines, MF=3, MT=2)  # elastic cross section cross-section
x, y = ENDF6.read_table(sec)

#plot quick
fig = plt.figure()
ax1 = fig.add_subplot(111)
plot(x, y)
ax1.set_yscale('log')
ax1.set_xscale('log')
xlabel('Neutron energy [eV]')
ylabel('Cross-section [barn]')
plt.savefig('figures/TENDL-17_random_70Ge_0000.png')
show()
```
